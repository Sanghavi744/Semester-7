```python
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
''' 
Hierarchical Clustering: Hierarchical clustering is a clustering 
technique used to group similar data points into nested clusters or 
a hierarchy of clusters. It can be represented as a tree-like 
structure (dendrogram) where the root represents a single cluster 
containing all data points, and the leaves represent individual data 
points. The algorithm works by iteratively merging or splitting 
clusters based on the similarity between data points. 
There are two main types of hierarchical clustering: 
agglomerative (bottom-up) and divisive (top-down).

'''
```

```python
''' 
Elbow Method: The elbow method is a technique used to determine the 
optimal number of clusters (k) for a clustering algorithm, such as 
K-Means. It involves running the clustering algorithm for a range of 
k values and plotting the within-cluster sum of squares (WCSS) or 
inertia against k. The point at which the plot forms an "elbow" or 
starts to level off is considered the optimal k value. The elbow point 
signifies the balance between minimizing intra-cluster variance and 
having a reasonable number of clusters.
'''
```

```python
''' 
K-Means Hierarchical Clustering: K-Means is a partitioning clustering 
technique that groups data points into non-overlapping clusters. 
In contrast, hierarchical clustering (agglomerative) creates a 
hierarchy of clusters where data points can belong to multiple levels 
of clusters. K-Means aims to partition data into a fixed number of 
clusters (k), while hierarchical clustering creates a hierarchical 
structure of clusters.
'''
```

```python
''' 
Difference Between Hierarchical Clustering and Normal Clustering:

1. Hierarchy vs. Fixed Clusters: Hierarchical clustering creates a 
     hierarchy of clusters, allowing data points to belong to multiple 
     levels of clusters, while normal clustering (e.g., K-Means) 
     partitions data into a fixed number of clusters.
2. Agglomerative vs. Partitioning: Hierarchical clustering can be 
    agglomerative (bottom-up) or divisive (top-down), merging or 
    splitting clusters based on similarity. Normal clustering, 
    like K-Means, is typically agglomerative and partitions data 
    into clusters.
3. Dendrogram vs. Cluster Assignment: Hierarchical clustering results 
    in a dendrogram, which visualizes the hierarchy of clusters. 
    In normal clustering, data points are directly assigned to a 
    specific cluster.
4. Flexibility vs. Fixed k: Hierarchical clustering is more flexible 
    in capturing hierarchical structures, while K-Means and other 
    normal clustering techniques require specifying the number of 
    clusters (k) in advance.
5. Interpretable vs. Non-overlapping: Hierarchical clustering is often 
    more interpretable as it shows the relationship between clusters 
    at different levels. Normal clustering techniques, like K-Means, 
    produce non-overlapping clusters that may not reveal such 
    relationships.
'''
```
